```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import scipy.stats as st
from pprint import pprint
import time
from scipy.stats import linregress
import reverse_geocoder as rg
import glob
```

```python
# Load in csv 
all_earthquakes = pd.read_csv("earthquake_data_1980-2020.csv")

all_earthquakes.head()
```

```python
year_quakes = all_earthquakes.groupby(["Year"]).count()
year_quakes
```

```python
# Plot the data
plt.plot(year_quakes["Mag"].index, year_quakes["Mag"], marker="o", color="blue")

# Format title and axis labels
plt.title("Earthquakes the last 40 years")
plt.xlabel("Time(Year)")
plt.ylabel("Count")
plt.grid(True)

# Save the graph
plt.savefig("images/Earthquake_Count.png")

#Display Graph
plt.show()
```
